## Notebook de prueba para el funcionamielto de Residual.py y gr√°fica de funciones

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import Residual as re
```

Creamos dos dataFrame indexados por fecha

```python
# Crear un rango de fechas
index = pd.date_range('2024-01-01', periods=5, freq='D')

# Crear el primer dataframe con una columna temporal
df1 = pd.DataFrame({'Temperatura': [22,  24,  24,  25,  28]}, index=index)

# Crear el segundo dataframe con una columna temporal diferente
df2 = pd.DataFrame({'Temperatura': [21,  22,  23,  24,  26]}, index=index)

# Mostrar los dataframes
print(df1)
print(df2)
```

Graficamos los datos df1 y df2

```python
re.viz_df1_df2(df1, df2, 'Comparaci√≥n de funciones', 'Fecha', 'Temperatura')
```

Calculo del residuo entre las dos funciones

```python
residual = re.calculate_residuals(df1, df2)
print(residual)
```

Graficamos las funciones y su residuo

```python
re.viz_df1_df2_residual(df1, df2, residual, 'Comparaci√≥n de funciones', 'Fecha', 'Temperatura')
```

Cargamos un dato como anomal√≠a conocida

```python
outliers_dates =  ["2024-01-04 "]

known_outliers = residual.loc[outliers_dates]
```

Graficamos las funciones, su residuo y la anomal√≠a cargada

```python
re.viz_df1_df2_residual_outk(df1, df2, residual, known_outliers,'Comparaci√≥n de funciones', 'Fecha', 'Temperatura')
```

A partir de una cota generamos los outliers como dato booleano

```python
cota = -2
outliers = re.binarizar_sobre_cota(residual, cota)
print (outliers)
```

Graficamos las funciones, el residuo, la anomal√≠a cargada y los outliers generados a partir de la cota

```python
re.viz_df1_df2_residual_outk_out(df1, df2, residual, known_outliers, outliers, 'Comparaci√≥n de funciones', 'Fecha', 'Temperatura')
```
